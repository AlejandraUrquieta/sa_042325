```python
def get_performance(choices):
    c3 = 0;
    c4 = 0;
    c5 = 0;
       
    for c in choices:
        if c == 3:
            c3+=1
        elif c == 4:
            c4+= 1
        elif c == 5:
            c5+=1;           
        
    p3 = (c3+c4+c5)/(len(choices)-2)
    p4 = (c4+c5)/(len(choices)-3)
    p5 = c5/(len(choices)-4)
    
    return p3, p4, p5;
```

```python
# changing number of trees - estimators

#>190 0.714 (0.080) - n_estimators=210, lr=0.02, max_depth=4

#>>210 0.733 (0.096) - n_estimators=210, lr=0.05, max_depth=4

#>140 0.689 (0.067) - n_estimators=140, lr=0.1, max_depth=4

#>50 0.674 (0.098) - n_estimators=50, lr=0.5, max_depth=4


def get_models_est():
	models = dict()
	# define number of trees to consider
	n_trees = [50,70,90,100,110,120,130,140,150,160,170,180,190,200,210,220,230,240,250,260,270]
	for n in n_trees:
		models[str(n)] = GradientBoostingClassifier(learning_rate=0.2,n_estimators=n, max_depth=4)
	return models
```
