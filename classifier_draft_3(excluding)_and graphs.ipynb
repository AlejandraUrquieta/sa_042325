```python
def get_performance(choices):
    c3 = 0;
    c4 = 0;
    c5 = 0;
       
    for c in choices:
        if c == 3:
            c3+=1
        elif c == 4:
            c4+= 1
        elif c == 5:
            c5+=1;           
        
    p3 = (c3+c4+c5)/(len(choices)-2)
    p4 = (c4+c5)/(len(choices)-3)
    p5 = c5/(len(choices)-4)
    
    return p3, p4, p5;
```
